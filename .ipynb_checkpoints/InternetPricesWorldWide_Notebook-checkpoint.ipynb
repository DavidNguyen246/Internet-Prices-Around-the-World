# Transformation :  Cleaning and merging of data 

```python
# rename the columns for internetusage_df
renamedinternetusage=internetusage_df.rename(columns={"country":"Country","incomeperperson":"Income Per Person","internetuserate":"Internet Use Rate","urbanrate":"Urban Rate"})
renamedinternetusage.head()
```

```python
# merge internetInfo and renamedinternentusage dataframes using "country"
firstMerge = pd.merge(internetInfo_df,renamedinternetusage, on = "Country")
firstMerge.head()
```

```python
# check how many rows merged together
len(firstMerge)
```

```python
#drop some columns that were not neccessary
firstMerge = firstMerge.drop(columns=['S.NO','Country code','Continental region','Urban Rate'])
firstMerge.head()
```

```python
# reordered dataframe by country in alphabetical order
sortedMerge = firstMerge.sort_values('Country')
len(sortedMerge)
```

```python
#renamed dataframe columns to be easier to read
renamedsortedMerge=sortedMerge.rename(columns={
    "NO. OF Internet Plans":"Internet Plans Available",
    "Average price of 1GB (USD)":"Average price of a GB in 2022",
    "Cheapest 1GB for 30 days (USD)":"Cheapest GB",
    "Most expensive 1GB (USD)":"Most expensive GB",
    "Average price of 1GB (USD  at the start of 2021)":"Average price of a GB in 2021",
    "Average price of 1GB (USD – at start of 2020)":"Average price of a GB in 2020",
    "Avg \n(Mbit/s)Ookla":"Average Mbit/s"
    })
renamedsortedMerge.head()
```

```python
# dropped any rows with 'NAN' and reset index 
cleaned_InternetDF = renamedsortedMerge.dropna().reset_index(drop=True)

#rounded any values that were decimals to two points 
rounded_InternetDF =cleaned_InternetDF.round(2)
rounded_InternetDF.head()
```

```python
# check how many rows are left 
len(rounded_InternetDF)
```

### Merging Broadband table with table showing prices

```python
#drop the columns we are not gonna use in database
mobileData = mobileData.drop(columns=['Mobile Speed Rank','Mobile Mbps','As of'])
mobileData.head()
```

```python
len(mobileData)
```

```python
internet_Broadband = pd.merge(rounded_InternetDF, mobileData, on = "Country")
internet_Broadband.head()
```

```python
#check how many rows are left 
len(internet_Broadband)
```

```python
#make sure we have no Null/NaN values
internet_Broadband.isnull().sum()
```

```python
# replacing the country name 

qoutes_removal['Country'] = qoutes_removal['Country'].str[1:]
qoutes_removal = qoutes_removal.replace('United States of America', 'United States')
```

```python
# merging the above dataset to our previous merged set. 
FinalMergeDF = pd.merge(internet_Broadband, qoutes_removal, on = 'Country')
FinalMergeDF = FinalMergeDF.rename(columns={"InternetPrice": "Internet Price of City (USD)", "Average Mbit/s": "Average National Internet Speed"})
FinalMergeDF
```

```python
FinalMergeDF = FinalMergeDF[['Country', 'City', 'Internet Price of City (USD)', 'Average price of a GB in 2020', 'Average price of a GB in 2021',
             'Average price of a GB in 2022', 'Internet Plans Available', 'Cheapest GB', 'Most expensive GB',
                            'Average National Internet Speed','Internet users', 'Internet Use Rate', 'Broadband Mbps']]
```

```python
engine = create_engine('postgresql://postgres:yourpassword@localhost/')
FinalMergeDF.to_sql('final_internet_data', engine)
```
